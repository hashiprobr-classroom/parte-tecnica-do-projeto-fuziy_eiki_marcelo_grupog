```python
def loading_data(data_dir):
    df = pd.read_csv(data_dir)
    df = df.loc[df["Stage"] != "Qualifying"]
    df = df.reset_index()
    team1 = []
    team2 = []
    for i in range(df.shape[0]):
        team1.append(df.iloc[i]["Team 1"].split("›")[0])
        team2.append(df.iloc[i]["Team 2"].split("›")[0])
    
    del df["Team 1"]
    del df["Team 2"]
    
    df["Team 1"] = team1
    df["Team 2"] = team2
    
    return df

df = loading_data("./champions/2015-16/champs.csv")
df
```
